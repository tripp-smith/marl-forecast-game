# marl-forecast-game Quickstart

This notebook demonstrates the public academic-facing API and a minimal deterministic run.

```python
from marl_forecast_game import GameEngine, SimulationConfig, demo_state

cfg = SimulationConfig(horizon=12, max_rounds=12, defense_model='ensemble')
game = GameEngine(cfg, seed=42)
out = game.run(demo_state(value=125.0, exogenous=0.2), rounds=12, disturbed=True)
out.forecasts[:5], out.targets[:5]
```
